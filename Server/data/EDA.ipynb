```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.cm import get_cmap
```

```python
for i in range(10):
    df = pd.read_csv("D_{:03d}.csv".format(i)).to_numpy()
    print(df)
    x = df[:, 1]
    y = df[:, 2]
    z = df[:, 3]
    s = df[:, 4]

    ind_surface = np.where(z == -.5)[0]
    plt.scatter(y[ind_surface], x[ind_surface], c=s[ind_surface], cmap=get_cmap("RdBu", 10), vmin=10, vmax=36)
    plt.colorbar()
    plt.show()

# df = pd.read_csv("D_002.csv").to_numpy()
# x = df[:, 1]
# y = df[:, 2]
# z = df[:, 3]
# s = df[:, 4]

# ind_surface = np.where(z == -.5)[0]
# plt.scatter(y[ind_surface], x[ind_surface], c=s[ind_surface], cmap=get_cmap("RdBu", 10), vmin=10, vmax=36)
# plt.colorbar()
# plt.show()


# df = pd.read_csv("D_003.csv").to_numpy()
# x = df[:, 1]
# y = df[:, 2]
# z = df[:, 3]
# s = df[:, 4]

# ind_surface = np.where(z == -.5)[0]
# plt.scatter(y[ind_surface], x[ind_surface], c=s[ind_surface], cmap=get_cmap("RdBu", 10), vmin=10, vmax=36)
# plt.colorbar()
# plt.show()
```
